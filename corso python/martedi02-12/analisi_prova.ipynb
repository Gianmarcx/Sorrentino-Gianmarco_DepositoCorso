```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt


data = load_iris() #caricamento dataset 

#Creazione di un dataframe con le caratteristiche
df = pd.DataFrame(data.data, columns=data.feature_names) 
df["target"] = data.target
df["species"] = df["target"].apply(lambda x : data.target_names[x])

print("Info sul dataset")
print(df.info())

print("Descrizione")
print(df.describe())

print("Prime 5 righe datset")
print(df.head())

# Conta quante osservazioni per specie
counts = df["species"].value_counts()


# Grafico a barre
plt.bar(counts.index, counts.values, color=["skyblue","salmon","lightgreen"])
plt.title("Numero di fiori per specie")
plt.xlabel("Specie")
plt.ylabel("Conteggio")
plt.show()
```
